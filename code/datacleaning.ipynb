```python
import json
import numpy as np
import pandas as pd
```

#### Read business dataset, and find the barbeque restaurant

```python
business = pd.read_json("business.json",lines = True)
```

```python
bbq = business[business["categories"].str.contains("Barbeque", na = False)]
bus_id = bbq["business_id"].tolist()
bbq.to_csv("BBQ.csv")
```

#### Read review dataset with chunks, filter the reviews of BBQ restaurant, and write them into a file

```python
review = pd.DataFrame()
```

```python
i = 0
for chunk in pd.read_json('review.json', lines=True, chunksize=20000):
#with pd.read_json("review.json", lines = True, chunksize=chunksize) as reader:
#for chunk in reader:
    print("Processing chunk ", i)
    review = pd.concat((review,chunk[chunk["business_id"].isin(bus_id)]))
    i += 1
```

```python
review.to_csv("review.csv")
```
